# 1) Instruction (for data cleaning)
    . Find all Null values in the dataset if there is any null value in any column, then fill it with the mean of that column

```python
car.isnull().sum()
```

```python
car['MSRP'] = car['MSRP'].replace(r'[\$,]', '', regex=True).astype(float)
car['Invoice'] = car['Invoice'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
car.fillna(car.mean(numeric_only=True), inplace=True)
```

# 3) Instruction (Filtering)
    . Show all the records where origin is Asia or Europe

```python
# method 1: using or operator
car[(car['Origin'] == 'Asia') | (car['Origin'] == 'Europe')]
```

```python
# method 2: using isin
car[car['Origin'].isin(['Asia','Europe'])]
```
